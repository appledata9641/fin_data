```python


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Black-Scholes公式計算選擇權價格
def black_scholes(S, K, T, r, sigma, option_type='call'):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    
    if option_type == 'call':
        price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    elif option_type == 'put':
        price = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    else:
        raise ValueError("option_type must be 'call' or 'put'")
    
    return price

# 設定參數
S0 = 100  # 標的資產價格
T = 1.0   # 到期時間（年）
r = 0.05  # 無風險利率
sigma = 0.2  # 波動率
option_type = 'put'  # 選擇權類型：'call' 或 'put'

# 履約價範圍
K_values = np.linspace(1, 300, 300)
option_prices = []

# 計算不同K下的選擇權價格
for K in K_values:
    price = black_scholes(S0, K, T, r, sigma, option_type)
    option_prices.append(price)

# 繪製圖形
plt.figure(figsize=(10, 6))
plt.plot(K_values, option_prices, label=f'{option_type.capitalize()} Option Price')
plt.xlabel('Strike Price (K)')
plt.ylabel('Option Price')
plt.title(f'Option Price vs. Strike Price ({option_type.capitalize()})')
plt.legend()
plt.grid(True)
plt.show()



```
